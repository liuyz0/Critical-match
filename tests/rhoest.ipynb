To show how $\rho_{\rm est}$ depends on $\sigma_Y/\mu_Y$ and $\sigma_C/\mu_C$.
$$
\rho = \sqrt{\frac{1}{1 + \frac{\sigma^2_Y}{\mu_Y^2} 
+ \frac{\sigma^2_Y\mu_C^2}{\sigma^2_C\mu_Y^2}}}.
$$

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt 
```

```python
CVC = np.linspace(0.5,5,100)
CVY = np.linspace(0,5,100)

X, Y = np.meshgrid(CVY, CVC)
Z = np.sqrt(1 / (1 + (1 + 1/Y**2)*X**2))
```

```python
plt.figure(figsize=(2.8, 2.4),constrained_layout=True)
#plt.imshow(Z, origin='lower',extent=(0,5,.5,5))
plt.contourf(X,Y,Z,10,cmap=mpl.colormaps["Blues"])
plt.clim(0,1)
plt.colorbar()
plt.ylabel(r'Consumption variation, $\sigma_C/\mu_C$')
plt.xlabel(r'Yield variation, $\sigma_Y/\mu_Y$')
plt.savefig('../figures/rhoest.pdf')
```
