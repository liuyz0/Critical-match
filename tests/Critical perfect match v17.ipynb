```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from utils import sample
```

```python
N = 16
M = 32
num_try = 100
sigY = 0
sigH_span = np.linspace(0,1,20)
NU = np.zeros((len(sigH_span), num_try))
for i in range(len(sigH_span)):
    sigH = sigH_span[i]
    for j in range(num_try):
        NU[i,j], _ = sample(N,M,0,sigY,sigH)
```

```python
FNC = np.mean((NU > 0).astype(float), -1)
std_FNC = np.std((NU > 0).astype(float), -1)/np.sqrt(num_try)
```

```python
plt.figure(figsize=(2.8, 2.1), constrained_layout=True)

plt.errorbar(sigH_span, FNC, std_FNC, ls='none', marker = 'v', c='k')

plt.ylim(0,1)
plt.xlim(0,1)
#plt.xscale('log')
plt.xlabel(r'High order interaction, $\sigma_H$')
plt.ylabel('Unstable fraction')
plt.savefig('../figures/test17.eps')
```
