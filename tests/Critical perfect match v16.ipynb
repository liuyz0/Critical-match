```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../src')
from utils import sample
```

```python
N = 16
M = 32
num_try = 100
sigY = 0
sigP_span = np.linspace(0,.4,20)
NU = np.zeros((len(sigP_span), num_try))
for i in range(len(sigP_span)):
    sigP = sigP_span[i]
    for j in range(num_try):
        NU[i,j], _ = sample(N,M,sigP,sigY,0)
```

```python
FNC = np.mean((NU > 0).astype(float), -1)
std_FNC = np.std((NU > 0).astype(float), -1)/np.sqrt(num_try)
```

```python
plt.figure(figsize=(2.8, 2.1), constrained_layout=True)

plt.errorbar(sigP_span, FNC, std_FNC, ls='none', marker = 's', c='k')

plt.ylim(0,1)
plt.xlim(0,.4)
#plt.xscale('log')
plt.xlabel(r'Production strength, $\sigma_P$')
plt.ylabel('Unstable fraction')
plt.savefig('../figures/test16.eps')
```
