```julia
using LinearAlgebra, Plots
```

Given the same $\rho$, change auto-regulation, to see change in four-leaf clover.

```julia
function sample(Ns,Nr,reg,rho)
    Tgc = randn(2,Ns,Nr)

    G = Tgc[1,:,:]
    C = rho * G + √(1 - rho^2) * Tgc[2,:,:]
    G = G .+ 3.0
    C = C .+ 3.0

    Js = zeros(Ns+Nr,Ns+Nr)
    Js[1:Ns,Ns+1:Ns+Nr] = G
    Js[Ns+1:Ns+Nr,1:Ns] = - C'
    Js[Ns+1:Ns+Nr,Ns+1:Ns+Nr] = - Matrix(I,Nr,Nr) * reg

    E_J = eigvals(Js) / √Nr
    E_LV = eigvals(-G * C') / Nr

    return E_J, E_LV
end
```

```julia
E_J = []
E_LV = []
for i in 1:5
    E_Jsamp, E_LVsamp = sample(32, 128, 1, 0.8) # let M greater
    append!(E_J, E_Jsamp)
    append!(E_LV, E_LVsamp)
end
```

The ellipse is between
$$
(\rho - \sqrt{r})(1 - \rho\sqrt{r})
$$
and 
$$
(\rho + \sqrt{r})(1 + \rho\sqrt{r})
$$
on real axis, while between
$$
\pm (1 - \rho^2)\sqrt{r}
$$
on the imaginary axis.

```julia
r = 32/128
rho = 0.8
```

```julia
scatter(real(E_J), imag(E_J), 
    size = (280, 210), 
    xlims = (-2,.5),
    ylims=(-1.5,1.5),
    legend = false,
    markeralpha = 0.2,
    markerstrokewidth = 0,
    xlabel = "Real part",
    ylabel = "Imaginary part")
scatter!(real(E_LV), imag(E_LV),
    markeralpha = 0.5,
    markerstrokewidth = 0)
plot!([0 for i = 1:10],[-1.5+0.34*(i-1) for i =1:10], color =:black, ls = :dash)
plot!((rho^2 + 1) * √r * cos.(LinRange(0,2π,100)) .- rho*(1 + r), (1 - rho^2) * √r * sin.(LinRange(0,2π,100)), color = :red)
```
