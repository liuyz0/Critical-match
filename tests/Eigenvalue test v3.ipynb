```python
import numpy as np
import matplotlib.pyplot as plt
```

Given the same $\rho$, change auto-regulation, to see change in four-leaf clover.

```python
# V matrix and Lambda matrix

def sample(Ns:int,
           Nr:int,
           reg:float,
           rho:float):
    
       Tgc = np.random.randn(2, Ns, Nr)

       G = Tgc[0]
       C = rho * Tgc[0] + np.sqrt(1-rho**2) * Tgc[1]
       G = G + 3.0
       C = C + 3.0

       E_LV = np.linalg.eigvals(- G @ C.T)/Nr

       return E_LV
```

```python
E_LV = [sample(16, 64, 1, 0.8) for i in range(6)] # let M greater
```

```python
plt.figure(figsize=(.933, 1.2), constrained_layout=True)
plt.plot([0 for i in range(10)], [-2 + i*4/9 for i in range(10)], 'k--', lw=1)
for i in range(4):
    plt.scatter(E_LV[i].real, E_LV[i].imag, s = 3)
plt.plot((.8**2 + 1) * np.sqrt(1/4) * np.cos(np.linspace(0,2*np.pi,100)) - .8*(1 + 1/4), (1 - .8**2) * np.sqrt(1/4) * np.sin(np.linspace(0,2*np.pi,100)), 'k',lw=1)
plt.axis('off')
plt.xlim(-2,.1)
plt.ylim(-.5,.5)
plt.savefig('../figures/fig3eigens-1.pdf')
```
