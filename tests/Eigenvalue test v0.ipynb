```python
import numpy as np
import matplotlib.pyplot as plt
```

Given the same $\rho$, change auto-regulation, to see change in four-leaf clover.

```python
# V matrix and Lambda matrix

def sample(Ns:int,
           Nr:int,
           reg:float,
           rho:float):
    
       Tgc = np.random.randn(2, Ns, Nr)

       G = Tgc[0]
       C = rho * Tgc[0] + np.sqrt(1-rho**2) * Tgc[1]
       G = G + 3.0
       C = C + 3.0

       Js = np.zeros((Ns+Nr,Ns+Nr))
       Js[0:Ns,Ns:Ns+Nr] = G
       Js[Ns:Ns+Nr,0:Ns] = - C.T
       Js[Ns:Ns+Nr,Ns:Ns+Nr] = - np.identity(Nr) * reg

       E_J = np.linalg.eigvals(Js)/np.sqrt(Nr)
       E_LV = np.linalg.eigvals(- G @ C.T)/Nr

       return E_J, E_LV
```

```python
E_J, E_LV = sample(64, 128, 1, 0.8) # let M greater
```

```python
plt.figure(figsize=(2.8, 2.1), constrained_layout=True)
plt.scatter(E_LV.real, E_LV.imag, c='b', s = 10, alpha=.5,marker='s')
plt.scatter(E_J.real, E_J.imag, c='r', s=10, alpha=.5)
plt.plot([0 for i in range(10)], [-2 + i*4/9 for i in range(10)], 'k--')
plt.xlim(-2,0.5)
plt.ylim(-2,2)
plt.savefig('../figures/fig3eigvals-0.pdf')
```

```python
E_J, E_LV = sample(64, 128, 16, 0.8) # let M greater
E_J = E_J[~np.isclose(E_J, np.zeros(np.shape(E_J)))]
```
