```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("../data/simu0.json")
results = json.load(file)
```

```python
data = np.zeros((20,10))
for i in range(20):
    for j in range(10):
        data[i,j] = results[str(i+1)][str(j+1)]['SF']
```

```python
mu = np.mean(data, -1)
sigma = np.std(data, -1)
```

```python
plt.figure(figsize=(2.8, 2.1), constrained_layout=True)
plt.errorbar(np.linspace(.05,1,20), mu, sigma, marker = 'o', ls='none')
plt.plot(np.ones(10)*np.sqrt(1/2), np.linspace(0,1,10), '--k')
plt.xlim(0,1)
plt.ylim(0,1)
plt.xlabel(r'Correlation, $\rho$')
plt.ylabel('Survival fraction')
```
